### 第三堂作業
### 第一題：

$\begin{equation}2x + y = 11 \end{equation}$

$\begin{equation}3x - y = 9 \end{equation}$

### 第二題：

$\begin{equation}\frac{x - 1}{y + 3} = \frac{1}{9}\end{equation}$

$\begin{equation}\frac{x + 3}{y - 5} = 5\end{equation}$

### 第三題：

$\begin{equation}4x + 3y = 4 \end{equation}$

$\begin{equation}2x + 2y - 2z = 0 \end{equation}$

$\begin{equation}5x + 3y + z = -2 \end{equation}$

```python
from sympy.solvers import solve
from sympy.core import sympify
from sympy import Symbol
import numpy as np
import matplotlib.pyplot as plt
x = Symbol('x')
y = Symbol('y')

# 第一題
print("第一題")
print("2x + y = 11")
print("3x - y = 9")
answer1 = solve([2*x + y - 11, 3*x - y -9])
x1=float(answer1[x])
y1=float(answer1[y])
print(f"x = {round(x1,0)}, y = {round(y1,0)}")

px = np.linspace(x1-10,x1+10,100)
py1 = np.array([11-2*i for i in px])
py2 = np.array([3*i-9 for i in px])
plt.plot(px, py1, color='blue',label='2x+y=11' )
plt.plot(px, py2, color="orange",label='3x-y=9' )
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()

# 第二題
print("\n第二題")
print('(x-1) / (y+3)= 1 / 9')
print("(x+3) / (y-5) = 5")
answer2 = solve(sympify("Eq((x-1)/(y+3), 1/9),Eq((x+3)/(y-5), 5)"))
x2=float(answer2[x])
y2=float(answer2[y])
print(f"x = {round(x2,0)}, y = {round(y2,0)}")

px = np.linspace(x2-10,x2+10,100)
py1 = np.array([9*i-12 for i in px])
py2 = np.array([(i+28)/5 for i in px])
plt.plot(px, py1, color='blue',label='(x-1) / (y+3)= 1 / 9' )
plt.plot(px, py2, color="orange",label='(x+3) / (y-5) = 5' )
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()

# 第三題
print("\n第三題")
print("4x + 3y = 4")
print("2x + 2y - 2z = 0")
print("5x + 3y + z = -2")
a = np.array([[4, 3, 0],[2,2,-2],[5,3,1]])
b = np.array([4, 0,-2])
x3,y3,z3 = np.linalg.solve(a,b)
print(f"x = {round(x3,0)}, y = {round(y3,0)}, z = {round(z3,0)}")

fig = plt.figure(figsize=(8,8))  
ax = plt.axes(projection='3d')
xx = np.linspace(x3-10,x3+10,100)
yy = np.linspace(y3-10,y3+10,100)
X, Y = np.meshgrid(xx, yy)
Z1 = np.zeros((X.shape[0],X.shape[1]))
Z2 = np.zeros((X.shape[0],X.shape[1]))
Z3 = np.zeros((X.shape[0],X.shape[1]))
for i in range(X.shape[0]):
    for j in range(X.shape[1]):
        Z2[i][j] = X[i][j] + Y[i][j]
        Z3[i][j] = -(5*X[i][j] + 3*Y[i][j]+2)

ax.plot_surface(X,Y,Z1,color = 'r',alpha=0.3)
ax.plot_surface(X,Y,Z2,color = 'g',alpha=0.3)
ax.plot_surface(X,Y,Z3,color = 'b',alpha=0.3)
ax.quiver(x3,y3,z3,x3+1,y3+1,z3+1,color='r',length=0.2)
ax.text(x3+1,y3+1,z3+1,f"x = {round(x3,0)}, y = {round(y3,0)}, z = {round(z3,0)}",color='k')
```
